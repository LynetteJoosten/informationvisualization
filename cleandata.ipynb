```python
import pandas as pd
df = pd.read_csv('metbuurten.csv')
```

```python
df['number_of_incidents'] = 1
df
```

```python
df['group'] = df['neighborhood_id_ams'] +'/'+df['incident_year_month'].map(lambda x: str(x))+'/'+df['incident_hour'].map(lambda x: str(x))
df2 = df.groupby('group').sum()[['number_of_incidents']]
```

```python
df2 = df2.reset_index()
```

```python
df2['code'] = df2['group'].map(lambda x: x.split('/')[0])
df2['year'] = df2['group'].map(lambda x: x.split('/')[1].split('-')[0])
df2['month'] = df2['group'].map(lambda x: x.split('/')[1].split('-')[1])
df2['hour'] = df2['group'].map(lambda x: x.split('/')[2])
```

```python
df2[['number_of_incidents','code','year','month','hour']].to_csv('perhourdata.csv')
```
